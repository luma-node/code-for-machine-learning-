```python
import pandas as pd
```

```python
dataset = pd.read_excel("/content/melb_data.xlsx")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.columns = dataset.columns.str.strip().str.lower()
dataset["councilarea"] = dataset["councilarea"].fillna(dataset["councilarea"].mode()[0])
```

```python
dataset.columns = dataset.columns.str.strip().str.lower()
dataset["propertycount"] = dataset["propertycount"].fillna(dataset["propertycount"].mode()[0])
```

```python
en_data = dataset[["councilarea", "propertycount"]]
en_data
```

```python
pd.get_dummies(en_data)
```

```python
pd.get_dummies(en_data).info()
```

```python
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop="first")
ar = ohe.fit_transform(en_data).toarray()
ar
```

```python
column_names = list(ohe.get_feature_names_out(["councilarea", "propertycount"]))
pd.DataFrame(ar, columns=column_names)
```
