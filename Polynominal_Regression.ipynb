```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/regression.csv")
```

```python
dataset.head(3)
```

```python
plt.scatter(dataset["Level"],dataset["Salary"])
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
x = dataset["Level"]
y = dataset["Salary"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)
x_reshaped = x.values.reshape(-1, 1)
pf.fit(x_reshaped)
x = pf.transform(x_reshaped)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
# y = m1x1+m2x2^2+c
# y = 1785.46221444*x1+(-6.27310327)+2^2+79065.893277865
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
prd = lr.predict(x)
```

```python
plt.scatter(dataset["Level"],dataset["Salary"])
plt.plot(dataset["Level"],prd,c='red')
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend("org","prd")
plt.show()
```
