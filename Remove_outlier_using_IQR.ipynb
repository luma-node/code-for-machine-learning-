```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel("/content/melb_data.xlsx")
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x="Rooms",data=dataset)
plt.show()
```

```python
dataset.shape
```

```python
q1 = dataset["Rooms"].quantile(0.25)
q3 = dataset["Rooms"].quantile(0.75)
q3
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (1.5*IQR)
max_range = q3 + (1.5*IQR)
min_range , max_range
```

```python
dataset
```

```python
new_dataset = dataset[dataset["Rooms"]<max_range]
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="Rooms",data=new_dataset)
plt.show()
```
