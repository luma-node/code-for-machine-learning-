```python
import pandas as pd
```

```python
df = pd.DataFrame({"Size":["m","l","xl","s","m","l","s","s","l","xl","m"]})
df.head()
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["Size"]])
```

```python
df["Size_en"] = oe.transform(df[["Size"]])
```

```python
df
```

```python
ord_data1 = {"s":10,"m":11,"x":12,"xl":13}
```

```python
df["Size_en_data"] = df["Size"].map(ord_data1)
```
