```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("/content/college_student_placement_dataset.csv")
```

```python
dataset.head(3)
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,lr.predict(x_test))
```

```python
sns.heatmap(dataset.corr(),annot=True)
plt.show()
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,lr.predict(x_test), average='weighted')*100
```

```python
from sklearn.metrics import recall_score
recall_score(y_test,lr.predict(x_test), average='weighted')*100
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,lr.predict(x_test), average='weighted')*100
```
