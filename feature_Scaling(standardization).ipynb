```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel("/content/melb_data.xlsx")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["Longtitude"] = dataset["Longtitude"].fillna(dataset["Longtitude"].mean())
```

```python
sns.distplot(dataset["Longtitude"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["Longtitude"]])
```

```python
ss.transform(dataset[["Longtitude"]])
```

```python
dataset["Longtitude_ss"] = pd.DataFrame(ss.transform(dataset[["Longtitude"]]), columns=["Longtitude_ss"])
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["Longtitude_ss"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["Longtitude"])

plt.show()
```
