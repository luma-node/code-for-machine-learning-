```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel("/content/melb_data.xlsx")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.boxplot(dataset["Distance"])
```

```python
sns.displot(dataset["Distance"])
```

```python
min_range = dataset["Distance"].mean() - (3*dataset["Distance"].std())
max_range = dataset["Distance"].mean() + (3*dataset["Distance"].std())
```

```python
min_range,max_range
```

```python
new_dataset = dataset[dataset["Distance"]<=max_range]
```

```python
new_dataset.shape
```

```python
sns.boxplot(x = "Distance",data = new_dataset)
```

```python
z_score = (dataset["Distance"] - dataset["Distance"].mean())/(dataset["Distance"].std())
```

```python
dataset["z_score"] = z_score
```

```python
dataset[dataset["z_score"]<3]
```
